```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
df = pd.read_csv('../screener/all.csv', index_col=0)
```

```python
sc = plt.scatter(df['d_sma05_sma20'], df['d_sma20_sma50'], c=df['expected'], vmin=-0.05, vmax=0.05, cmap=cm.seismic, s=1)
plt.colorbar(sc)
plt.show()
```

```python
tmp = df.query('0.05 <= expected')
sc = plt.scatter(tmp['d_sma05_sma20'], tmp['d_sma20_sma50'], c=tmp['expected'], vmin=0.05, vmax=0.1, cmap=cm.seismic, s=0.1)
plt.colorbar(sc)
plt.xlim([-0.2, 0.2])
plt.ylim([-0.2, 0.2])
plt.show()
```
